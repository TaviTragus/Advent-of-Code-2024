```python
import pandas as pd
```

```python
#importing data!
url_numbers = 'https://raw.githubusercontent.com/TaviTragus/Advent-of-Code-2024/refs/heads/main/Day1/Puzzle1_Data.csv'
numbers = pd.read_csv(url_numbers,dtype={'Gc':'category'})
numbers
```

```python
number_1 = numbers["first set"].sort_values()
number_2 = numbers[" second set"].sort_values()
```

```python
number_1
```

```python
number_2
```

```python
num1 = number_1.values.tolist()
num2 = number_2.values.tolist()
```

```python
result = 0
index = list(range(0,1000))
for i in index:
  running_total = abs(num1[i]-num2[i])
  result += running_total
result
```
